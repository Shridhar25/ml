```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score 
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.30, random_state=4)
t=iris.target
```

```python
C = 1.0

# For Linear Kernel
linear = svm.SVC(kernel='linear').fit(x_train, y_train)
y_pred = linear.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)*100
print("Accuracy for SVM with Linear Kernel:",accuracy)

# For Polynomial Kernel
polynomial = svm.SVC(kernel='poly', degree=4).fit(x_train, y_train)
y_pred = polynomial.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)*100
print("Accuracy for SVM with Polynomial Kernel:",accuracy)

# For Sigmoid Kernel
sigmoid = svm.SVC(kernel='sigmoid').fit(x_train, y_train)
y_pred = sigmoid.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)*100
print("Accuracy for SVM with Sigmoid Kernel:",accuracy)

# For RBF Kernel
rbf = svm.SVC(kernel='rbf', gamma=1).fit(x_train, y_train)
y_pred = rbf.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)*100
print("Accuracy for SVM with RBF Kernel:",accuracy)
```

```python
# Plotting Linear Kernel 
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

plt.subplot(1, 1, 1)
linear1 = svm.SVC(kernel='linear').fit(X, t)
Z = linear1.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

plt.scatter(X[:, 0], X[:, 1], c=t, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Linear Kernel')
plt.show()
```
